```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plot the sketch of relation between $a^{\pm}$ and $t$

```python
aMinus = np.arange(-1.5, -1 + 0.001, 0.001)
aPlus = np.arange(1, 1.5, 0.001)
tMinus1 = aMinus + np.sqrt(aMinus**2 - 1)
tMinus2 = aMinus - np.sqrt(aMinus**2 - 1)
tPlus1 = aPlus + np.sqrt(aPlus**2 - 1)
tPlus2 = aPlus - np.sqrt(aPlus**2 - 1)
plt.plot(aMinus, tMinus1, c='b', label='$t = a^- + \sqrt{\left(a^-\\right)^2 - 1}, t = a^+ - \sqrt{\left(a^+\\right)^2 - 1}$')
plt.plot(aMinus, tMinus2, c='orange', label='$t = a^- - \sqrt{\left(a^-\\right)^2 - 1}, t = a^+ + \sqrt{\left(a^+\\right)^2 - 1}$')
plt.plot(aPlus, tPlus1, c='orange')
plt.plot(aPlus, tPlus2, c='b')
plt.hlines((-1, 1), -1.5, 1.5, linestyles='dashed', colors='r')
plt.xlabel('$a^{\pm}$')
plt.ylabel('$t$')
plt.legend()
plt.show()
plt.close()
```
